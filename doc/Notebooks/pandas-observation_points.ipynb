```python
import sys
sys.path.append("C:/Program Files/DHI/2019/FEFLOW 7.2/python")
import ifm_contrib as ifm 
ifm.forceLicense("Viewer")
doc = ifm.loadDocument("./example_data/example_2D.dac")
```

The `doc.c.mesh.gdf` object provides functions to obtain the location of observation points as a GeoDataframe (2D and 3D Layered only).

```python
doc.c.obs.gdf.getGeoDatframe()
```

Plotting is most usefull when combined with other GeoPandas objects of the model.

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.axis('equal')
ax1 = plt.gca()  # get current axis
doc.c.mesh.gdf.getGeoDatframe().plot(ax=ax1)
doc.c.obs.gdf.getGeoDatframe().plot(ax=ax1, color="red")
# add labels
for label, row in doc.c.obs.gdf.getGeoDatframe().iterrows():
    plt.annotate(s=label, 
                 xy=(row['x'], row['y']),
                 horizontalalignment='center')
```
